# Get Prediction from the Model

```python
import tensorflow as tf

import matplotlib.pyplot as plt
import numpy as np

from sde.sde_learning_network import \
    (
        SDEIdentification,
        ModelBuilder,
        SDEApproximationNetwork,
    )

from sde.experiment_reports import \
(
    sample_data,
    plot_results_functions,
    generate_results,
    plot_results_1d
)
from sde.thesis_experiments import \
(
    loadSIRInfo,
    loadSIRInfo1D
)

%load_ext autoreload
%autoreload 2
```

```python
# notebook parameters
random_seed = 1
step_size = 1e-1  # 5e-2 # step size
n_pts = 5000        # number of points

n_layers = 5
n_dim_per_layer = 25

n_dimensions = 1

ACTIVATIONS = tf.nn.tanh
VALIDATION_SPLIT = .2
BATCH_SIZE = 512
N_EPOCHS = 40

# only diagonal diffusivity matrix (does not matter since we are in 1D)
diffusivity_type = "diagonal"

tf.random.set_seed(random_seed)
x_data,y_data,x_data_test,y_data_test = loadSIRInfo1D()

step_sizes = np.zeros((x_data.shape[0],)) + step_size
# define the neural network model we will use for identification
encoder = ModelBuilder.define_gaussian_process(
                                        n_input_dimensions=n_dimensions,
                                        n_output_dimensions=n_dimensions,
                                        n_layers=n_layers,
                                        n_dim_per_layer=n_dim_per_layer,
                                        name="GP",
                                        activation=ACTIVATIONS,
                                        diffusivity_type=diffusivity_type)
model = SDEApproximationNetwork(sde_model=encoder, method="euler")
model.compile(optimizer=tf.keras.optimizers.Adamax())

sde_i = SDEIdentification(model=model)
hist = sde_i.train_model(x_data, y_data, step_size=step_sizes,
                         validation_split=VALIDATION_SPLIT, n_epochs=N_EPOCHS, batch_size = BATCH_SIZE)
fig, hist_axes = plt.subplots(1, 1, figsize=(4, 2))
hist_axes.clear()
hist_axes.plot(hist.history["loss"], label='loss')
hist_axes.plot(hist.history["val_loss"], label='validation')
hist_axes.legend()

def get_estimate(current_infected_rate):

    current_data = np.array([[current_infected_rate]])
    T_steps = 2
    N_iterates = 1
    rng = np.random.default_rng()
    time_steps, paths_network = \
        generate_results(sde_i.drift_diffusivity,
                         step_size, current_data, rng,
                         T_steps=T_steps, N_iterates=N_iterates);
    check = False
    # Check if estimate is reasonable for prediction: bigger than 0.01, not too much bigger (limit:0.3),
    while not check:
        if paths_network[0][1] < current_infected_rate or paths_network[0][1] > current_infected_rate + 0.3:
            time_steps, paths_network = \
        generate_results(sde_i.drift_diffusivity,
                         step_size, current_data, rng,
                         T_steps=T_steps, N_iterates=N_iterates);
        else:
            estimate = paths_network[0][1]
            check = True
            
    return estimate
get_estimate(current_infected_rate=0.04)
```

# Scenario 1

```python
def show_graph(node_labels = {},edge_labels={},title='None'):
    G = nx.MultiDiGraph()
    node_names = ['Market', 'Work', 'Transport']
    my_color_map = ['orange', 'yellow', 'lightblue']
    G.add_nodes_from(node_names)
    pos = nx.spring_layout(G)
    pos[node_names[0]] = [0, 0]
    pos[node_names[1]] = [1.0, 0]
    pos[node_names[2]] = [0.5,0.8]
    node_sizes = [5000, 5000, 5000]
    G.add_edges_from([(node_names[0], node_names[2])])
    G.add_edges_from([(node_names[0], node_names[1])])
    G.add_edges_from([(node_names[1], node_names[2])])
    
    options = {
        'node_color': my_color_map,
        'node_size': node_sizes,
        'width': 6,
        'arrowstyle': '-|>',
        'arrowsize': 5,
        'pos': pos,
        'arrows':False,
        'with_labels':True,
        'font_size':9,
        'font_color':'black',
        'labels':node_labels
    }

    plt.figure(figsize=(8, 8))
    draw_nodes = nx.draw_networkx(G,  **options)
    #edges = nx.draw_networkx_edges(G, pos)
    if title:
        plt.title(title, fontsize=24, color='black')
    
    #nx.draw_networkx_labels(G, pos, font_size=36, )
    draw_edge_labels = nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red',font_size=12)
    plt.axis('off')
    plt.show()
    
# move_info = {    
#     ('Market', 'Work'):[1,50],
#     ('Market', 'Transport'):[1,50],
#     ('Work', 'Market'):[1,50],
#     ('Work', 'Transport'):[1,50],
#     ('Transport', 'Market'):[4,180],
#     ('Transport', 'Work'):[0,20],
# }

def refresh_edge_labels(node_names,move_info):
    return {
        (node_names[0], node_names[1]): 
        str(move_info[(node_names[0], node_names[1])][0])+'/'+
        str(move_info[(node_names[0], node_names[1])][1])+'-->'+'\n'
        +'<--'+
        str(move_info[(node_names[1], node_names[0])][0])+'/'+
        str(move_info[(node_names[1], node_names[0])][1]),
        
        (node_names[0], node_names[2]): 
        str(move_info[(node_names[0], node_names[2])][0])+'/'+
        str(move_info[(node_names[0], node_names[2])][1])+'-->'
        +'\n'
        +'<--'+
        str(move_info[(node_names[2], node_names[0])][0])+'/'+
        str(move_info[(node_names[2], node_names[0])][1]),
        
        (node_names[2], node_names[1]): 
        str(move_info[(node_names[2], node_names[1])][0])+'/'+
        str(move_info[(node_names[2], node_names[1])][1])+'-->'
        +'\n'
        +'<--'+
        str(move_info[(node_names[1], node_names[2])][0])+'/'+
        str(move_info[(node_names[1], node_names[2])][1])
    }
# node_info = [[12,100],[16,100],[22,100]]
def nodes_after_infection(node_info):
    
    for i in range(len(node_info)):
        current_infected_rate = node_info[i][0]/node_info[i][1]
        new_infected = get_estimate(current_infected_rate)
        node_info[i][0] = math.floor(node_info[i][1]*new_infected)
    return node_info
        
def nodes_after_movement(node_info,move_info,node_names_index):
    for move in move_info:
        source_index = node_names_index[move[0]]
        dest_index = node_names_index[move[1]]
        move_amount_infected = move_info[move][0]
        move_amount_total = move_info[move][1]
        node_info[source_index][0] -=move_amount_infected
        node_info[source_index][1] -=move_amount_total
        node_info[dest_index][0] +=move_amount_infected
        node_info[dest_index][1] +=move_amount_total
    return node_info
def refresh_node_labels(node_info,node_names):
    node_labels = {
        node_names[0]:'Market\n '+str(node_info[0][0])+'\\'+str(node_info[0][1]),
        node_names[1]:'Work/School\n '+str(node_info[1][0])+'\\'+str(node_info[1][1]),
        node_names[2]:'Transportation\n '+str(node_info[2][0])+'\\'+str(node_info[2][1])

    }
    return node_labels
def total_infected(node_info):
    result = 0
    for node in node_info:
        result +=node[0]
    return result
    
```

# Ilk Scenario Day 1 Started

```python
###initial_information\
# node_names = ['Market', 'Work', 'Transport']
node_info = [[2,100],[7,700],[2,200]]
#node_info = nodes_after_infection(node_info)
node_names = ['Market', 'Work', 'Transport']

node_labels = refresh_node_labels(node_info,node_names)
show_graph(node_labels=node_labels,title='1. Day Start State Total Infected :'+str(total_infected(node_info)))

# First Move:
node_names = ['Market', 'Work', 'Transport']
node_names_index={'Market':0, 'Work':1, 'Transport':2}
move_info = {    
    ('Market', 'Work'):[1,50],
    ('Market', 'Transport'):[1,50],
    ('Work', 'Market'):[1,50],
    ('Work', 'Transport'):[1,50],
    ('Transport', 'Market'):[4,180],
    ('Transport', 'Work'):[0,20],
}

edge_labels = refresh_edge_labels(node_names,move_info)
#show_graph(node_labels,edge_labels,title='Iteration 1')
node_info = nodes_after_movement(node_info,move_info,node_names_index)
node_labels = refresh_node_labels(node_info,node_names)
show_graph(node_labels=node_labels,edge_labels=edge_labels,title='1. State After Moving')
    
    
    
node_info = nodes_after_infection(node_info)
node_labels = refresh_node_labels(node_info,node_names)

show_graph(node_labels=node_labels,title='2. State Total Infected :'+str(total_infected(node_info))) 
```
